```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup

url = 'https://spongebob.fandom.com/wiki/List_of_transcripts'
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
soup.findAll('a')
```

```python

anchors = [a for a in (td.find('a') for td in soup.findAll('td')) if a]
anchors = anchors[1::2] # this gives me just the transcripts. I need start from 0 if I want plot synopses too
anchors[0]
len(anchors)
```

```python

for i in range(0,len(anchors)):
    if anchors[i].has_attr('href'):
        link = anchors[i]['href']
        actual_link = 'https://spongebob.fandom.com'+link 
        file_name = './transcripts/spongebob_ep'+str(i)+'.html'
        urllib.request.urlretrieve(actual_link,file_name)
        file = open(file_name,'r')
        new_file = [n.replace('<dl><dd>','') for n in file.readlines() if n.startswith('<dl')]
        file.close()
        new_name = open('./transcripts/spongebob_ep'+str(i)+'_redacted.html','w')
        for k in range(len(new_file)):
            new_name.write(new_file[k]+'\n')
        new_name.close() 
    time.sleep(1)
```
